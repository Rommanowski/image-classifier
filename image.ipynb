```python
import jsonlines
import numpy as np
import cairo
import matplotlib
from PIL import Image
import os

%matplotlib notebook

class FileMaker:

    def add_points(self, obj):
        scratch = np.zeros((256, 256))
        for j in obj['drawing']:
            for k in range(len(j[0])):
                x = j[0][k]
                y = j[1][k]
                scratch[y][x] = 255
        return scratch
    
    def paint_lines(self, obj):
        line_diameter = 20
        surface = cairo.ImageSurface(cairo.FORMAT_ARGB32, 256, 256)
        cr = cairo.Context(surface)
        cr.set_antialias(cairo.ANTIALIAS_BEST)
        cr.set_line_cap(cairo.LINE_CAP_ROUND)
        cr.set_line_join(cairo.LINE_JOIN_ROUND)
        cr.set_line_width(line_diameter)
        cr.set_line_width(3)
        
        for j in obj['drawing']:
            cr.set_source_rgb(0, 0, 0)
            cr.paint()
            for k in range(0, len(j[0]) - 1):
                cr.set_source_rgb(255, 255, 255)
                x = j[0][k]
                y = j[1][k]
                cr.move_to(x, y)
                x1 = j[0][(k + 1)]
                y1 = j[1][(k + 1)]
                cr.line_to(x1, y1)
        cr.stroke()          
    
        data = surface.get_data()
        newscratch = np.copy(np.asarray(data)[::4])
        lines = newscratch.reshape(256, 256)
        return lines
    
    def merge_dots_and_lines(self, scratch, lines):
        scratch[:, :256] = lines
        scratch = scratch / 255
        final = np.where(scratch < 0.1, 1, 0)
        return final
    
    def save_training_img(self, final, name):
        os.makedirs(os.path.dirname(name), exist_ok=True)
        matplotlib.image.imsave(name, final, cmap='gray')

    def makeImages(self, file_name):
        full_file_name = 'full_simplified_' + file_name + '.ndjson'
        with open(full_file_name, encoding='utf-8') as file:
            allScratches = jsonlines.Reader(file)
            for i, obj in enumerate(allScratches):
                if i > 100:
                    break
                dots = self.add_points(obj)
                lines = self.paint_lines(obj)
                final = self.merge_dots_and_lines(dots, lines)
                image_name = obj['key_id'] + '.jpg'
                self.save_training_img(final, 'images/' + file_name + '/' + image_name)
                print(image_name)

# Usage:
# fm = FileMaker()
# fm.makeImages("yoga")
```

```python
filemaker = FileMaker()
```

```python
filemaker.makeImages('yoga')
```
